# Numpy Arrays

NumPy arrays are the main way we will use Numpy throughout the course. A NumPy array is a multidimensional array of objects all of the same type. In memory, it is an object which points to a block of memory, keeps track of the type of data stored in that memory, keeps track of how many dimensions there are and how large each one is, and - importantly - the spacing between elements along each axis.
Numpy arrays essentially come in two flavors: vectors and matrices. Vectors are strictly 1-d arrays and matrices are 2-d (but you should note a matrix can still have only one row or one column).

Let's begin our introduction by exploring how to create NumPy arrays.

## NumPy's ndarray vs Python's list object size
Before we jump straight into the practical applications, let’s compare the memory consumption between an ndarray and a list.

```python
import numpy as np
import sys

py_arr = [1,2,3,4,5,6]
numpy_arr = np.array([1,2,3,4,5,6])

sizeof_py_arr = sys.getsizeof(1) * len(py_arr)           # Size = 168 in bytes
sizeof_numpy_arr = numpy_arr.itemsize * numpy_arr.size   # Size = 24 in bytes

print(sizeof_py_arr, sizeof_numpy_arr)
```

## Storing floating points
Further optimization of NumPy’s arrays are also possible if one knows the number of individual data. NumPy’s arrays also save a lot of space when storing floating point values, in comparison with Python’s lists.

```python
py_arr = [1.1, 1.2,1.3424242, 
          1.436654646546, 454353.1232131, 
          32434353.324324, 1234.2321]
          
numpy_arr = np.array([1.1, 1.2,1.3424242, 
                      1.436654646546, 454353.1232131, 
                      32434353.324324, 1234.2321])

sizeof_py_arr = sys.getsizeof(12345.232343) * len(py_arr) # Size  = 168
sizeof_numpy_arr = numpy_arr.itemsize * numpy_arr.size    # Size = 56
print(sizeof_py_arr,sizeof_numpy_arr)
```

## Numerical Ranges
For those familiar with MATLAB, NumPy offers a very useful attribute to create evenly spaced range of values. The three attributes are:
1. numpy.arange(start, stop, step, dtype)
2. numpy.linspace(start, stop, num, endpoint, retstep, dtype)
3. numpy.logspace(start, stop, num, endpoint, base, dtype)

The first one, np.arange(), generates evenly spaced values in a given range. The second one, np.linspace(), generates a list of values based on the number of evenly spaced values between a certain interval. The last one, np.logspace() generates evenly spaced values in the logarithmic scale.

### arange
Return evenly spaced values within a given interval.

```python
np.arange(0,10)
```

```python
np.arange(0,11,2)
```

### linspace
Return evenly spaced numbers over a specified interval. Helps in creating bins or graphs for which we need equidistant points.

```python
np.linspace(0,10,3)
```

```python
np.linspace(0,10,50)
```

```python
np.linspace(0,10,50, retstep = True) #by default retstep = False
```

```python
np.logspace(10, 20, 8, base = 2)
```

## Using reshape and ndim

```python
a = np.array([[1,2,3],[4,5,6]])
print(a, 'number of dimensions: ',a.ndim)
print('Using reshape: ')
a_reshape = a.reshape(1,6)
print(a_reshape, 'number of dimensions: ',a_reshape.ndim)
print(' ')
```

## Resizing array using shape
Shape is an attribute that arrays have (not a method):

```python
a = np.array([[1,2,3],[4,5,6]])
print("Before resizing: ")
print(a)
print("Shape of array: ", a.shape)
print(' ')
a.shape=(3,2)
print('Resizing ndarray: ')
print(a)
print("Shape of array: ", a.shape)
print(' ')
```
