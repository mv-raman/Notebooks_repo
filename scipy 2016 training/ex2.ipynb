```python
from scipy import sparse
rnd=np.random.RandomState(seed=123)
x=rnd.uniform(low=0.0,high=1.0,size=(10,5))
print(x)
```

```python
x[x<0.7]=0
print(x)
```

```python
x_csr=sparse.csr_matrix(x)
print(x_csr)
```

```python
print(x_csr.toarray())
```

```python
x_lil=sparse.lil_matrix((5,5))
for i,j in np.random.randint(0,5,(15,2)):
    x_lil[i,j]=i+j
print(x_lil)
print(type(x_lil))
```

```python
x_dense=x_lil.toarray()
print(x_dense)
print(type(x_dense))
```

```python
x_csr=x_lil.tocsr()
print(x_csr)
print(type(x_csr))
```

```python
x=np.linspace(1,12,100)
y=x[:,np.newaxis]
im=y*np.sin(x)*np.cos(y)
print(im.shape)
plt.imshow(im)
```

```python
plt.contour(im)
```

```python
from mpl_toolkits.mplot3d import Axes3D
ax=plt.axes(projection="3d")
x_axis,y_axis=np.meshgrid(x,y.ravel())
ax.plot_surface(x_axis,y_axis,im,cmap=plt.cm.jet, cstride=2, rstride=2, linewidth=0);
```
