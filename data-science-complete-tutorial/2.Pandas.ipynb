```python
left=pd.DataFrame({'key':[1,2,3],
                  'A':['a','b','c']})
```

```python
right=pd.DataFrame({'key':[1,2,3],
                  'B':['d','e','f']})
```

```python
left.merge(right,on='key',how='left')
```

```python
left.merge(right,on='key')
```

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                     index=['K0', 'K1', 'K2'])

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                      'D': ['D0', 'D2', 'D3']},
                     index=['K0', 'K2', 'K3'])
```

```python
left.join(right)
```
