```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.DataFrame({'A':range(10),'B':range(5,15)})
```

```python
df.plot.line(title='Info')
```

```python
plt.plot(df.A,label='A')
plt.plot(df.B,label='B')
plt.xlabel('index')
plt.ylabel('values')
plt.title('info')
plt.legend
```

```python
plt.scatter(df.A,df.B,s=10)
```
