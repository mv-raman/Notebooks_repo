```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier,export_graphviz,ExtraTreeClassifier
```

```python
iris=load_iris()
```

```python
iris.data[:5]
```

```python
iris.target[:5]
```

```python
dt=DecisionTreeClassifier(criterion='entropy')
```

```python
from sklearn.model_selection import train_test_split
trainX,testX,trainY,testY=train_test_split(iris.data,iris.target)
```

```python
dt.fit(trainX,trainY)
```

```python
export_graphviz(dt,'dt.tree')
```

```python
dt.predict(testX)
```

```python
from sklearn.datasets import make_blobs
X,Y = make_blobs(n_features=2, n_samples=1000, cluster_std=.8, centers=4, random_state=6)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(X,Y)
```

```python
plot_step=0.2
x_min,x_max=X[:,0].min()-1,X[:,0].max()+1
y_min,y_max=X[:,1].min()-1,X[:,1].max()+1
xx,yy=np.meshgrid(np.arange(x_min,x_max,plot_step),np.arange(y_min,y_max,plot_step))
```

```python
outcome=dt.predict(np.c_[xx.ravel(),yy.ravel()])
```

```python
xx.shape,yy.shape
```

```python
plt.scatter(X[:,0],X[:,1],c=Y,cmap='viridis')
plt.scatter(xx.ravel(),yy.ravel(),c=outcome,alpha=1,s=1,cmap='viridis')
plt.show()
```
