```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from sklearn.learning_curve import validation_curve
from sklearn.datasets import load_svmlight_files
from sklearn.cross_validation import StratifiedKFold
from sklearn.datasets import make_classification
from xgboost.sklearn import XGBClassifier
from scipy.sparse import vstack

# reproducibility
seed = 123
np.random.seed(seed)
```

```python
X, y = make_classification(n_samples=1000, n_features=20, n_informative=8, n_redundant=3, n_repeated=2, random_state=seed)
```

```python
cv=StratifiedKFold(y,n_folds=10,shuffle=True,random_state=True)
```

```python
default_params={
    'objective':'binary:logistic',
    'max_depth':1,
    'learning_rate':0.3,
    'silent':1
}
n_estimator_range=np.linspace(1,200,10).astype('int')
train_score,test_score=validation_curve(XGBClassifier(**default_params),X,y,param_name='n_estimators',
                                       param_range=n_estimator_range,cv=cv,scoring='accuracy')
```

```python
train_score,test_score
```

```python
train_scores_mean = np.mean(train_score, axis=1)
train_scores_std = np.std(train_score, axis=1)
test_scores_mean = np.mean(test_score, axis=1)
test_scores_std = np.std(test_score, axis=1)
plt.plot(n_estimator_range,np.mean(train_score,axis=1),label='training score')
plt.plot(n_estimator_range,np.mean(test_score,axis=1),label='cv score')
plt.title("Validation Curve with XGBoost (eta = 0.3)")
plt.fill_between(n_estimator_range, 
                 train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, 
                 alpha=0.2, color="r")

plt.fill_between(n_estimator_range,
                 test_scores_mean - test_scores_std,
                 test_scores_mean + test_scores_std,
                 alpha=0.2, color="g")
plt.xlabel("number of trees")
plt.ylabel("Accuracy")
plt.legend(loc='best')
```

```python
i=np.argmax(np.mean(test_score,axis=1))
print('best cross validation score is {0:.2f} obtained in {1} trees'.format(np.mean(test_score,axis=1)[i],n_estimator_range[i]))
```
